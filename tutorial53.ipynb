```python
import pandas as pd

df1 = pd.read_csv("HistoricalPrices.csv")
df1head = df1.head(10)
df = df1head.rename(columns = {' Open': 'Open', ' High': 'High', ' Low': 'Low', ' Close': 'Close'})
df
```

```python
import matplotlib.pyplot as plt
plt.plot(df['Date'], df['Close'])
plt.title('Close Price')
plt.xlabel('Date')
plt.ylabel('Close')
plt.show()
```

```python
plt.plot(df['Date'], df['Open'], label = 'Open')
plt.plot(df['Date'], df['Close'], label = 'Close')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Open and Close Price')
plt.legend()
plt.show()
```

```python
# vertical bar
plt.bar(df['Date'], df['Open'])
plt.xlabel('Date')
plt.ylabel('Open')
plt.title('Open Price')
plt.show()
```

```python
# horizontal bar
plt.barh(df['Date'], df['Open'])
plt.xlabel('Open')
plt.ylabel('Date')
plt.title('Open Price')
plt.show()
```

```python
plt.scatter(df['Open'], df['Close'])
plt.xlabel('Open')
plt.ylabel('Close')
plt.title('Open vs Close')
plt.show()
```

```python
# box plot
plt.boxplot(df['Open'])
plt.xlabel('Open')
plt.ylabel('Close')
plt.title('Open Price')
plt.show()
```

```python
import seaborn as sns

sns.histplot(df['Open'])
plt.xlabel('Open')
plt.ylabel('Close')
plt.title('Open Price')
plt.show()
```
